### Il teorema del limite centrale afferma che la somma (o la media) di n variabili indipendenti aventi identica distribuzione è una variabile che si distribuisce normalmente qualsiasi sia la tipologia di distribuzione di partenza. 
Attenzione: 
Il teorema del limite centrale è un teorema limite. Questo vuole dire che l’approssimazione ad una gaussiana migliora all’aumentare del valore n. A bassi valori di n, infatti, l’approssimazione si verifica se la distribuzione di partenza è una distribuzione normale. Per distribuzioni di partenza molto asimmetriche, l’approssimazione non vale ed è richiesto solitamente un valore di n maggiore di 30.

L’applicazione più semplice da comprendere e più utilizzata del teorema del limite centrale è quella che riguarda la media campionaria. Si consideri di avere dei dati con una distribuzione di partenza qualsiasi e di campionare da questo set casualmente n valori. Si calcoli la media di tali valori.

La media altro non è che la somma di n variabili identicamente distribuite (come richiesto dal teorema sopra riportato) divisa per il valore n. Ripetendo questo campionamento altre volte, si ottengono altrettante medie. Secondo il teorema del limite centrale tali medie si distribuiscono normalmente.

La distribuzione della media campionaria sarà una normale con una media ed una deviazione standard come sopra riportate. Il fatto di poter trasformare qualsiasi distribuzione in una gaussiana con una deviazione standard ridotta rispetto alla distribuzione originale è un’arma molto potente in fase di analisi statistiche.

```python
# GAUSSIANA: distribuzione NORMALE STANDARDIZZATA
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
x=np.linspace(-6,6,50)  # 50 valori x da -6 a 6
# valori y tali che MEDIA = 0 e DEVIAZIONE STANDARD = 1 (scarto quadratico medio) (VARIANZA è il quadrato dello scarto)
y=np.exp(-x*x/2)/np.sqrt(2*np.pi)
plt.title("la funzione gaussiana standardizzata")
plt.plot(x, y, marker = "o", color = 'red')
plt.xlabel("X") 
plt.ylabel("Y")
plt.show()
# visualizzazione della media ponderata (vicina a 0!)
print ("Media = ",np.average(x,None,y))
```

```python
# generazione di N = 1000 valori con MEDIA 0.0 e DEVIAZIONE STANDARD 1.0 distribuiti secondo la FUNZIONE GAUSSIANA
x = np.random.normal( 0.0, 1.0, 1000 )
print ("Media = ",np.mean(x))
print ("Deviazione standard = ",np.std(x))
#  CREAZIONE DEL RELATIVO ISTOGRAMMA
Nc = 50  # numero di intervalli (= numero barre dell'istogramma)
pdf, bins = np.histogram( x, Nc, density = True ) # density = True : area sotto la curva è 1 (densità di probabilità)
print("Valori Asse X:")
print (bins)  # valori dell'asse x
print("Valori Asse Y:")
print (pdf)   # valori dell'asse y (frequenze = altezza delle barre)
nextr = len( bins )
print (nextr) # numero di barre
dx = bins[1] - bins[0]
print (dx)  # larghezza delle barre
plt.xlabel( r"$x$" )
plt.ylabel( "PDF" )
plt.bar( bins[:nextr-1], pdf, dx, color = "g" )
plt.show()
```

```python
# SIMULAZIONE DEL LANCIO DI 50 DADI (VARIABILI QUANTITATIVE DISCRETE) per 9 volte
# Studio della distribuzione della media delle 9 volte (array di 50 elementi)

# rv=random var : size numeri casuali interi compresi tra 1 e (high-1)
# array di 100 interi casuali compresi tra 1 e 6
dadi1=np.random.randint(1,high=7,size=50)
dadi2=np.random.randint(1,high=7,size=50)
dadi3=np.random.randint(1,high=7,size=50)
dadi4=np.random.randint(1,high=7,size=50)
dadi5=np.random.randint(1,high=7,size=50)
dadi6=np.random.randint(1,high=7,size=50)
dadi7=np.random.randint(1,high=7,size=50)
dadi8=np.random.randint(1,high=7,size=50)
dadi9=np.random.randint(1,high=7,size=50)

#Estraggo i valori unici ed il numero di occorrenze per dadi1
uniqueVal, occourC=np.unique(dadi1,return_counts=True)
freqrel=occourC/occourC.sum()
plt.bar(uniqueVal,freqrel,color='yellow')
plt.title("primi 50 lanci")
plt.show()  # per non avere tutto su un solo grafico

#Estraggo i valori unici ed il numero di occorrenze per dadi9
uniqueVal, occourC=np.unique(dadi9,return_counts=True)
freqrel=occourC/occourC.sum()
plt.bar(uniqueVal,freqrel)
plt.title("ultimi 50 lanci")
plt.show()  # per non avere tutto su un solo grafico

#Estraggo i valori unici ed il numero di occorrenze per la media delle 9 volte
rv = (dadi1+dadi2+dadi3+dadi4+dadi5+dadi6+dadi7+dadi8+dadi9)/9
uniqueVal, occourC=np.unique(rv,return_counts=True)
freqrel=occourC/occourC.sum()
#Per plottare le frequenze relative usero' : x= uniqueVal, y=freqrel
# print ("Valori unici       :",uniqueVal)
# print ("Numero occorrenze  :",occourC)
# print ("Frequenze relative :",freqrel)
plt.bar(uniqueVal,freqrel,color='red')
# media con linea continua
plt.axvline(np.mean(rv), c='navy')
# mediana con linea tratteggiata
plt.axvline(np.median(rv), c='navy', linestyle='--')
# primo e terzo quartile con linea puntinata
plt.axvline(np.quantile(rv,0.25),c='navy', linestyle=':')
plt.axvline(np.quantile(rv,0.75),c='navy', linestyle=':')
plt.title("media dei 9 lanci: ANDAMENTO GAUSSIANO!")
plt.show()
```
