```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
X = np.array([8,6,9,4,3,3,2,7,7]) 
media = np.mean(X)
devstd = np.std(X)
min = np.amin(X)
max = np.amax(X)
print ('media:',media)
print ('devstd:',devstd)
plt.plot(X)
```

```python
# definisco la funzione che trasforma la distribuzione di valori di partenza in una distribuzione con la stessa forma
# ma con valori compresi tra 0 e 1
def norm(x):
   return (x - min) / (max -min)
Xnorm = norm(X)
display(Xnorm)
print ('media:',np.mean(Xnorm))
print ('devstd:',np.std(Xnorm))
plt.plot(Xnorm)
```

```python
# con map e lambda function
l=list(map(lambda x:(x - min) / (max -min),X))
print(l)
plt.plot(l)
```

```python
# definisco la funzione che trasforma la distribuzione di valori di partenza in una distribuzione con la stessa forma
# ma con MEDIA=0 e DEVIAZIONE STANDARD=1
def standard(x):
  return  (x - media) / devstd
Xstd = standard(X)
display(Xstd)
print ('media:',np.mean(Xstd))
print ('devstd:',np.std(Xstd))
plt.plot(Xstd)
```
